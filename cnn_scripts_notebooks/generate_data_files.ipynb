```python
import os
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
train_path = '/home/carolinemckee/split_train/images/'
val_path = '/home/carolinemckee/split_val/images/'
test_path = '/home/carolinemckee/split_val/images/'
```

```python
train_list = os.listdir(train_path) 
val_list = os.listdir(val_path)
```

```python
with open('yolo_train.txt','w') as out_file:
    for t in train_list:
        out_file.write(train_path + t + '\n')
with open('yolo_validation.txt','w') as out_file:
    for t in val_list:
        out_file.write(val_path + t + '\n')
```

```python
test = [0,0.34716796875,0.90478515625,0.1826171875,0.1728515625]
test = [item*1024 for item in test]
img = cv2.imread(val_path + 'P0841__1__1374___0.png',cv2.IMREAD_COLOR)
print(img.shape)
print((int(test[1]+0.5*test[3]),int(test[2]+0.5*test[4])))
cv2.rectangle(img,(int(test[1]-0.5*test[3]),int(test[2]-0.5*test[4])),(int(test[1]+0.5*test[3]),int(test[2]+0.5*test[4])),(0,0,255),3)
plt.imshow(img)
plt.show()
```
